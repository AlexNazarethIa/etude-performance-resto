## 1. Chargement

```python
import pandas as pd
import  seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("restaurant_data_clean.csv")
df.head()
```

ce dataset présente 3 type de varibles:
**variables catégorielles** : Categorie, Type, Méthode Paiement
**variables numériques** : Prix, Quantité, Total Commande
**variables temporelles** : Date, Année, Mois, Jour, Numero Semaine

## 2. les tendances de ventes

```python
counts=df["Categorie"].value_counts()
sns.barplot(x=counts.index,y=counts.values)
```

Les ventes sont réparties de manière équilibrée entre les catégories, aucune catégorie ne domine fortement.

Analyser, pour chaque catégorie, la répartition des types les plus vendus

```python
# Grouper les données par catégorie et type, et sommer les quantités
type_counts = df.groupby(["Categorie", "Type"])["Quantité"].sum().reset_index()

# Trier par catégorie puis par quantité décroissante
type_counts = type_counts.sort_values(["Categorie", "Quantité"], ascending=[True, False])

# Visualisation avec Seaborn
plt.figure(figsize=(16, 6))
sns.barplot(
    data=type_counts,
    x="Type",
    y="Quantité",
    hue="Categorie"
)

# Rotation des labels pour meilleure lisibilité
plt.xticks(rotation=40)
plt.xlabel("Type d'article")
plt.ylabel("Quantité vendue")
plt.title("Quantité vendue par type et catégorie")
plt.legend(title="Catégorie")
plt.tight_layout()
plt.show()
```

À l'intérieur de chaque catégorie, certains types dominent clairement en termes de quantité vendue. Par exemple, dans Main Dishes, certains plats ont des volumes beaucoup plus élevés que les autres. On observe aussi une forte variabilité entre les types d'une même catégorie, ce qui indique que tous les produits ne contribuent pas de manière égale aux ventes.\n\n\nLes performances commerciales sont davantage influencées par le type de produit que par la catégorie elle-même.

## 3. Tendances de Ventes Temporelles

```python
# Dashboard temporel
plt.figure(figsize=(15, 12))
plt.subplot(2,2,1)
sales_year = df.groupby('Année')['Total Commande'].sum()
sales_year.plot(kind='bar')
plt.title('CA par Année')
plt.ylabel('CA (€)')
plt.subplot(2,2,2)
qty_year = df.groupby('Année')['Quantité'].sum()
qty_year.plot(kind='bar')
plt.title('Quantité par Année')
plt.ylabel('Quantité')
plt.subplot(2,2,3)
sales_month = df.groupby('Mois')['Total Commande'].sum()
sales_month.plot(kind='line', marker='o')
plt.title('CA par Mois')
plt.ylabel('CA (€)')
plt.xticks(rotation=45)
plt.subplot(2,2,4)
sales_day = df.groupby('Jour')['Total Commande'].sum()
sales_day.plot(kind='bar')
plt.title('CA par Jour')
plt.ylabel('CA (€)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

le CA a baissé de 2022 à 2023 car la quantité a baissée.

Nous allons donc chercher à comprendre pour la quantité a baissée.

```python

# Créer la figure
plt.figure(figsize=(18, 12))

# Boucle pour chaque catégorie
for i, cat in enumerate(categories, start=1):
    plt.subplot(2, 3, i)  # 2 lignes, 3 colonnes
    # Filtrer les données pour la catégorie
    cat_data = df[df["Categorie"] == cat].groupby(["Année", "Mois"])["Quantité"].sum().reset_index()
    
    # Graphique ligne pour voir l'évolution par mois
    sns.lineplot(
        data=cat_data,
        x="Mois",
        y="Quantité",
        hue="Année",
        marker="o",
        palette="Set2"
    )
    
    plt.title(cat)
    plt.xticks(range(1,13))
    plt.xlabel("Mois")
    plt.ylabel("Quantité")
    plt.legend(title="Année")

plt.tight_layout()
plt.show()
```

Certaines périodes montrent une baisse marquée :

- Janvier : 2272 → 2172 (-100)
- Juin : 2360 → 2075 (-285)
- Octobre : 2438 → 2147 (-291)

la baisse n’est pas uniforme, certaines périodes sont plus touchées (milieu et fin d’année).

Des périodes spécifiques ont contribué fortement à la baisse du CA, probablement liées à la demande ou à des événements particuliers.

```python
# Quantité par jour de la semaine
quant_par_jour = df.groupby("Jour")["Quantité"].sum()
print(quant_par_jour)
```

la répartition est assez uniforme entre les jours, aucun jour spécifique ne cause la baisse.

## 4. Produits les Plus Vendu et Moins  Vendus

```python
top_5 = df.groupby('Type')['Quantité'].sum().sort_values(ascending=False).head(5)
print("Top 5 produits les plus vendus :\n", top_5)
print()
bottom_5 = df.groupby('Type')['Quantité'].sum().sort_values(ascending=True).head(5)
print("Top 5 produits les moins vendus :\n", bottom_5)
```

il faudrait plus travailler sur les produits les moins vendus . Essayer de  voir le périodes  où  ces ventes ont diminuées.

```python
# Exemple pour un produit spécifique : "Pizza"
 
products=[ "Nachos Grande", "Iced Tea","Onion Rings","Vegetarian Platter","Garlic Bread"]
plt.figure(figsize=(15,12))
for i,prod in enumerate(products,start=1):
    plt.subplot(2,3,i)
    prod_data=df[df["Type"]==prod].groupby(['Année','Mois'])["Quantité"].sum().reset_index()
    sns.lineplot(data=prod_data, x='Mois', y='Quantité', hue='Année', marker='o',palette="Set1")
    plt.title(prod)
    plt.xticks(range(1,13))
    plt.xlabel("Mois")
    plt.ylabel("Quantité")
    plt.legend(title="Année")
plt.tight_layout()
plt.show()
```

**Insights Produits :**\n- Grilled Chicken et Pasta Alfredo dominent (quantité & CA)\n- Salmon haut de gamme performant\n- Focus sur ces best-sellers
